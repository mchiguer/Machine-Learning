# The Iris dataset

The Iris dataset has been widely used to introduce basic machine learning concepts and methods. It consists of $N=150$ samples, $50$ for each of $C=3$ classes, represented by vectors of $D=4$ homogeneous real features. One of the classes is linearly separable from the rest, but the other two are not linearly separable. Although today it is considered a "toy" dataset, it is still very useful for introducing basic concepts and methods.

First I import some standard and sklearn libraries:

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

Reading the Iris dataset:

```python
iris = load_iris()  # Charge le dataset Iris.  
print(dir(iris))  # Liste les attributs/méthodes de l'objet Iris.  
X = iris.data  # Données : mesures des pétales / Chaque ligne représente un échantillon / Chaque colonne représente une caractéristique (feature) mesurée
y = iris.target  # Cibles : étiquettes des espèces (0, 1, 2) / premier element de y corresponf a la classe de la fleur de la premiere ligne X 
fn = iris.feature_names  # Noms des caractéristiques (features)
cn = iris.target_names  # Noms des classes (espèces)  


print(iris.DESCR)  # Affiche une description détaillée du dataset Iris (format texte) 



#iris.data iris.target .... on decouvre les noms de dir(..dataset), ce ne sont pas des fonctions generales 
```

<p style="page-break-after:always;"></p>

I convert the corpus into a pandas dataframe to facilitate its description:

```python
data = pd.DataFrame(data=X, columns=fn) #ca cree une data frame(un peu comme un tab excel avec la matrice X et elle donne aux C les noms de fn)
data['species'] = pd.Series(iris.target_names[y], dtype='category') #Ajoute une nouvelle colonne species au DataFrame data
data #ca print la data
```

<p style="page-break-after:always;"></p>

Let's look at some basic statistics:

```python
data.describe() #on applique sur un dataFrame et nous donne les infos sur chaque caracteristique
```

I check that I have $50$ samples of each class:

```python
data.groupby('species', observed=False).size() #Ca crée des ss-groupes, un pour chaque espèce et .size calcule la taille de chaque groupe
```

<p style="page-break-after:always;"></p>

Since I have few features, it's a good idea to make a scatter matrix plot:

```python
sns.pairplot(data, hue="species", height = 1.4, palette = 'colorblind')    #chaque graphique montre la relation entre deux caractéristiques d ou vient le mot pairplot

#sns pr seaborn et cette biblio génère des graphiques visuellement attrayants par défaut
```

#The Digits dataset

Like iris, Digits can be considered a "toy" dataset. However, compared to iris, Digits represents a jump of complexity due to the greater number of classes, $C=10$, samples, $N=1797$, and dimension of feature vectors, $D=64$. In addition, digits addresses one of the main perceptual tasks of machine learning: optical character recognition (OCR) and, more specifically, handwritten digit recognition. Although handwritten digit recognition has been considered a "solved" task since the 1990s, image classification in general remains a complex task of great academic and commercial interest. So the relative simplicity of Digits is very convenient as an introductory task to image classification.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits() #charge le dataset
print(dir(digits)) #ce que contient le dataset
print(digits.target)
print(digits.DESCR) #description du dataset
```

<p style="page-break-after:always;"></p>

Let's see the first $10$ images:

```python
_, axes = plt.subplots(nrows=1, ncols=10, figsize=(16, 16))  # Crée une figure avec 1 ligne et 10 colonnes d'axes (10 sous-graphiques)
for ax, image, label in zip(axes, digits.images, digits.target):  # Itère sur les axes, les images et les labels du dataset
    ax.set_axis_off()  # Désactive l'affichage des axes
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="none")  # Affiche chaque image en niveaux de gris sans interpolation
    ax.set_title("Sample {!s}".format(label))  # Ajoute un titre à chaque image avec le label correspondant
```

# The Olivetti dataset

Olivetti contains $N=400$ face images of $C=40$ people, with 10 images per person. The images were acquired at different times, varying the lighting, facial expression (eyes closed or not; smiling or not) and facial details (with or without glasses). All of them are normalized to $64\times 64$ grayscale pixels between $0$ and $1$; that is, each image can be seen as a vector of $D=4096$ dimensions of real features in $[0, 1]$. People are identified with an integer class label from $0$ to $39$.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
```

```python
orl = fetch_olivetti_faces() #permet de charger un dataset d'images de visages
print(dir(orl))
print(orl.DESCR)
```

Let's see the first image of each person:

```python
nrows, ncols = 4, 10  
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(18, 18*nrows/ncols), constrained_layout=True)  #Crée une figure avec une grille de 4x10 sous-graphes, la taille de la figure est ajustée pour être proportionnelle à la grille
for c in np.arange(0, 40):  # Itère de 0 à 39, donc sur 40 visages
    ax = axes.flat[c]  # Pour chaque sous-graphe (ax)
    ax.set_axis_off()  #on désactive les axes,
    ax.set_title(f"Person {c}")    #on définit un titre avec le numéro de la personne (c)
    ax.imshow(orl.images[10*c], cmap=plt.cm.gray, interpolation="none")  # Affiche l'image du visage de la personne c (indexée à 10*c) en niveaux de gris, sans interpolation des pixels
```

<p style="page-break-after:always;"></p>

Let's see some people's $10$ images:

```python
cc = [0, 9, 21, 36]
nrows, ncols = len(cc), 10
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(18, 18*nrows/ncols), constrained_layout=True)
for i, c in enumerate(cc):
    for j in np.arange(0, 10):
        ax = axes.flat[10*i+j]; ax.set_axis_off(); ax.set_title(f"Pers. {c} face {j}")
        ax.imshow(orl.images[10*c+j], cmap=plt.cm.gray, interpolation="none")
```
